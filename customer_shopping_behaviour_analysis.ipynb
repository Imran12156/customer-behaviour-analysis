```python
df.isnull().sum()
```

```python
df['Review Rating'] =df.groupby('Category')['Review Rating'].transform(lambda x:  x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_') 
df = df.rename(columns ={'purchase amount_(usd)':'purchase amount'})
```

```python
#create a colummn age_group 
labels = ['Young Adult','Adult','Middle Aged','Senior']
df['age_group'] =pd.qcut(df['age'],q=4,labels =labels)
```

```python
#create column purchase_frequency_days

frequency_mapping  = {
    'Fortnightly':14,
    'Weekly': 7,
    'Monthly' : 30,
    'Quarterly' :90,
    'Bi-Weekly' :14,
    'Annually' :365,
    'Every 3 Months':90
}
df['purchase_freguency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
                      
```
